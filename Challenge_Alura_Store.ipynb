# Importación de datos



````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
````

#1. Análisis de facturación



````python
# Sumar todos los valores de la columna Precio para obtener el Ingreso Total de cada tienda

ingreso_total_tienda1 = tienda1['Precio'].sum()
ingreso_total_tienda2 = tienda2['Precio'].sum()
ingreso_total_tienda3 = tienda3['Precio'].sum()
ingreso_total_tienda4 = tienda4['Precio'].sum()
````

````python
# Ordenar los resultados para obtener cuáles tiendas tienen el mayor y el menor ingreso

ingresos_totales_tiendas = [ingreso_total_tienda1, ingreso_total_tienda2, ingreso_total_tienda3, ingreso_total_tienda4]
nombres_tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

combinacion_ingresos_nombres = sorted(zip(nombres_tiendas, ingresos_totales_tiendas))
````

````python
# Imprimir los resultados obtenidos

print(f'''El ingreso total por cada tienda es:
Tienda 1 → ${ingreso_total_tienda1:,.2f}
Tienda 2 → ${ingreso_total_tienda2:,.2f}
Tienda 3 → ${ingreso_total_tienda3:,.2f}
Tienda 4 → ${ingreso_total_tienda4:,.2f}

El amyor ingreso total fue proveniente de la {combinacion_ingresos_nombres[0][0]} con un valor de ${combinacion_ingresos_nombres[0][1]:,.2f}
El menor ingreso total fue proveniente de la {combinacion_ingresos_nombres[-1][0]} con un valor de ${combinacion_ingresos_nombres[-1][1]:,.2f}''')
````

# 2. Ventas por categoría

````python
# Obtener categorías de productos de cada tienda

categorias_tienda1 = tienda1['Categoría del Producto'].unique()
categorias_tienda2 = tienda2['Categoría del Producto'].unique()
categorias_tienda3 = tienda3['Categoría del Producto'].unique()
categorias_tienda4 = tienda4['Categoría del Producto'].unique()
````

````python
# Utilizar el método set() para crear un conjunto de categorías únicas y posteriormente, ordenarlas para utilizar los índices.

categorias_conjunto = set(categorias_tienda1) & set(categorias_tienda2) & set(categorias_tienda3) & set(categorias_tienda4)
categorias = sorted(categorias_conjunto)

print(f"Categorías de productos: {categorias}")
````

````python
# Por cada tienda:
# 1. Inicializar un contador por cada Categoría del Producto
# 2. Ejecutar un código con lazo for para contar cada ítem de la columna Categoría del Producto y así sumar a los contadores iniciales
````

# 3. Calificación promedio de la tienda
